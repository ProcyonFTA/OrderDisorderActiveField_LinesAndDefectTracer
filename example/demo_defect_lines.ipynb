```julia
using Pkg
Pkg.resolve()
Pkg.activate("..")

using DefectComplexScalarOrderParameter
DCSOP = DefectComplexScalarOrderParameter
```

```julia
# import complex scalar field 
ϕ = DCSOP.import_phi_field("data/example_snapshot_nrch.h5");

# create cube for the surface charges on the lattice
nx,ny,nz = size(ϕ)
defect_map= zeros(Float64,nx-1,ny-1,nz-1,3)

# compute charge on all vertices
@time DCSOP.cpu_defect_cube!(ϕ,defect_map=defect_map)

# extract defect points
points = DCSOP.defect_points(defect_map;nx=nx,ny=ny,nz=nz)
```

```julia
# trace lines from the points

@time inline, infos_line, extremity_direction = DCSOP.line_collection(points)
```

```julia
#export the data to be read on Paraview

@time DCSOP.write_lines_vtk(points,inline,"data/example_lines.vtk")
```
